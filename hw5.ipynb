```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
```

```python
dataset = pd.read_csv('texts_dataset.csv', delimiter=' ')
print(f'dataset shape: {dataset.shape}')
dataset["initial_text"] = dataset["texts"]
dataset.head()
```

Фильтрация текста

```python
only_letters = lambda s: ''.join(list(filter(lambda c: c.isalpha() or c == ' ', s)))
dataset["texts"] = [only_letters(comm.lower()) for comm in dataset["texts"]]
```

Количество уникальных слов

```python
len(' '.join(dataset["texts"]).split(' '))
```

```python
from itertools import chain
from typing import List
from collections import Counter

def build_dictionary(s: List[str]) -> Counter:
    return Counter(list(chain.from_iterable([text.split() for text in s])))
```

Удаление стоп-слов

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
eng_stopwords = set(stopwords.words('english'))
dataset["texts"] = [' '.join(list(filter((lambda word: word not in eng_stopwords), text.split()))) for text in dataset["texts"].tolist()]
```

Лемматизация

```python
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
dataset["texts"] = [' '.join(list(map(lemmatizer.lemmatize, text.split()))) for text in dataset["texts"]]
dictionary = build_dictionary(dataset["texts"])
```

```python
import random

numbers = [None] * len(dataset["sentiment"])
for i in range (len(dataset["sentiment"])):
    if dataset["sentiment"][i] == "Positive":
        numbers[i] = 1
    if dataset["sentiment"][i] == "Neutral":
        numbers[i] = 0
    if dataset["sentiment"][i] == "Negative":
        numbers[i] = -1   
dataset["sentiment_in_numbers"] = numbers    
random.shuffle(dataset["texts"])
dataset.head()
```
